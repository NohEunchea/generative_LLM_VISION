```python
from bs4 import BeautifulSoup
import requests
```

```python
main_url = "https://namu.wiki/w/%EB%8D%B0%EC%9D%B4%EB%B8%8C%20%EB%8D%94%20%EB%8B%A4%EC%9D%B4%EB%B2%84"
response = requests.get(main_url)
web_page = response.text
web_page
```

```python
soup = BeautifulSoup(web_page, 'html.parser')
soup
```

```python
temp = soup.select("h3>span")
temp
```

```python
main_dic = {}
for item in temp:
    id_ = item.get("id")
    link = item.select_one("a").get("href")

    ## 필요한 정보는 /w로 시작
    if link.startswith("/w"):
        main_dic[id_] = "https://namu.wiki"+link
print(main_dic)
```

```python
test = main_dic["등장인물"]
response = requests.get(test)
web_page = response.text
soup = BeautifulSoup(web_page, 'html.parser')
res = soup.select("div.wRLRAq2n")
for x in res:
    print(x.text)
```

```python
import time
for name, url in main_dic.items():
    response = requests.get(url)
    web_page = response.text
    soup = BeautifulSoup(web_page, 'html.parser')
    contents = soup.select("div.wRLRAq2n")
    result = f"{name}\n"
    for content in contents:
        result += content.text
    with open(f"{name}.txt", "w", encoding="utf-8-sig") as file:
        file.write(result)
    print(name, '저장 완료')
    time.sleep(2)
```
